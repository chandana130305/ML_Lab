math library

```python
import math
```

```python
num = 9
angle_degrees = 45
print ("Square root : ",math.sqrt(num))
print ("Factorial : ",math.factorial(5))
print ("Logarithm base 10 : ",math.log10(100))
print ("Sine of 45 degrees : ",math.sin(math.radians(angle_degrees)))
print ("Value of PI : ",math.pi)
```

```python
num1=4.3
num2 = 9.7
num3=-2.8
num4=5.0
print("Ceil of",num1," is : ",math.ceil(num1))
print("Ceil of",num2," is : ",math.ceil(num2))
print("Ceil of",num3," is : ",math.ceil(num3))
print("Ceil of",num4," is : ",math.ceil(num4))
```

scipy library

```python
from scipy import linalg
import numpy as np

two_d_array = np.array([[4, 5], [3, 2]])

det_value = linalg.det(two_d_array)

print("Determinant of the matrix:", det_value)
```
