```python
#numpy basics
import numpy as np

a = np.array([1, 2, 3])   # Create a rank 1 array
print(type(a))
print(a.shape)            # Prints "(3,)"
print(a[0], a[1], a[2])   # Prints "1 2 3"
a[0] = 5                  # Change an element of the array
print(a)                  # Prints "[5, 2, 3]"
b = np.array([[1,2,3],[4,5,6]])    # Create a rank 2 array
print(b.shape)                     # Prints "(2, 3)"
print(b[0, 0], b[0, 1], b[1, 0])   # Prints "1 2 4"

```

scipy library

```python
#SciPy (Scientific Python) is another free and open-source Python library
#It’s extensively used for scientific and technical computations, because it extends NumPy and provides many user-friendly
#and efficient routines for scientific calculations.1.	SciPy is built in top of the NumPy
#	SciPy is a fully-feature diversion of Linear Algebra while Numpy contains only a few features.
#	Most new Data Science features are available in Scipy rather than Numpy.
from scipy import linalg
import numpy as np  # Import NumPy for array creation

# Define a 2×2 square matrix
two_d_array = np.array([[4, 5], [3, 2]])

# Compute the determinant using SciPy
det_value = linalg.det(two_d_array)

# Print the determinant value
print("Determinant of the matrix:", det_value)

```
