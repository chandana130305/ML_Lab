```python
import statistics
```

```python
score = [25, 45, 67, 89, 34, 25, 67, 89, 23, 90, 99, 88, 77, 67, 66, 55]
```

```python
mean_score = statistics.mean(score)
median_score = statistics.median(score)
mode_score = statistics.mode(score)
variance_score = statistics.variance(score)
standard_deviation = statistics.stdev(score)
```

```python
#print output
print(f"MEAN : {mean_score}")
print(f"MEDIAN : {median_score}")
print(f"MODE : {mode_score}")
print(f"VARIANCE : {variance_score}")
print(f"STANDARD DEVIATION : {standard_deviation}")
```
