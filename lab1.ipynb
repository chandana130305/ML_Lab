```python
import statistics
```

```python
score = [25, 45, 67, 89, 34, 25, 67, 89, 23, 90, 99, 88, 77, 67, 66, 55]
```

```python
mean_value = statistics.mean(score)
median_value = statistics.median(score)
mode_value = statistics.mode(score)
variance_value = statistics.variance(score)
standard_deviation = statistics.stdev(score)
```

```python
#print output
print(f"MEAN : {mean_value}")
print(f"MEDIAN : {median_value}")
print(f"MODE : {mode_value}")
print(f"VARIANCE : {variance_value        }")
print(f"STANDARD DEVIATION : {standard_deviation}")
```
