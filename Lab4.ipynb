**Experiment-4 :**
*Write a Python program to implement Simple Linear Regression*

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])   #no. of hours studied
y = np.array([40, 50, 60, 70, 80])   #marks obtained
n = len(x)  #no.of students
```

```python
#Step-1 : Calculate Sums
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x * x)
```

```python
#Step-2 : Calculate slope(m) and intercept(c)
m = (n * sum_xy - sum_x * sum_y)/(n * sum_x2 - sum_x * sum_x)
c = (sum_y-m*sum_x)/n
```

```python
#Step-3 : Predict marks using the line y = m*x+c
y_pred = m*x+c
```

```python
#Step-4 : Show result
print(f"Slope(m) = {m:.2f}")
print(f"Intercept(c) = {c:.2f}")
print(f"Predicted marks = {y_pred}")
```

```python
#Step-5 : Draw Picture
plt.scatter(x,y,color = 'blue')
plt.plot(x,y_pred,color = 'red')
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```
